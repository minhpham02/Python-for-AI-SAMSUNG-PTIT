```python
import pandas as pd
import numpy as np
```

Hiển thị toàn bộ dữ liệu

```python
df = pd.read_csv('P4AI_BT1.csv')
display(df)
```

Điền các giá trị bị thiếu của dữ liệu theo phương pháp trung bình với các trường có giá trị số và theo phương pháp tần số cao nhất với giá trị rời rạc

```python
df1 = df.copy()
df1.isna().sum()
```

```python
# Điền các giá trị bị thiếu của dữ liệu theo phương pháp trung bình

df1['sepal.length'].fillna(round(df['sepal.length'].mean()), inplace = True)
df1['sepal.width'].fillna(round(df['sepal.width'].mean()), inplace = True)
df1['petal.length'].fillna(round(df['petal.length'].mean()), inplace = True)
df1['petal.width'].fillna(round(df['petal.width'].mean()), inplace = True)

# Điền giá trị còn thiếu theo phương pháp tần số cao nhất

df1['variety'].fillna(df['variety'].mode()[0], inplace = True)

display(df1.isna().sum())

df1
```

Hiển thị các bản ghi có sepal.length > 5 và sepal.width > 3

```python
df_filtered = df1[(df1['sepal.length'] > 5) & (df1['sepal.width'] > 3)]
display(df_filtered)
```

Chuẩn hóa dữ liệu theo min max

```python
df2 = df1.copy()

sepal_length_max = df2['sepal.length'].max()
sepal_length_min = df2['sepal.length'].min()

sepal_width_max = df2['sepal.width'].max()
sepal_width_min = df2['sepal.width'].min()

petal_length_max = df2['petal.length'].max()
petal_length_min = df2['petal.length'].min()

petal_width_max = df2['petal.width'].max()
petal_width_min = df2['petal.width'].min()

# Chuẩn hóa

df2['sepal.length'] = (df2['sepal.length'] - sepal_length_min) / (sepal_length_max - sepal_length_min)
df2['sepal.width'] = (df2['sepal.width'] - sepal_width_min) / (sepal_width_max - sepal_width_min)
df2['petal.length'] = (df2['petal.length'] - petal_length_min) / (petal_length_max - petal_length_min)
df2['petal.width'] = (df2['petal.width'] - petal_width_min) / (petal_width_max - petal_width_min)

df2
```

Liên tục hóa thuộc tính variety

```python
df3 = df2.copy()

# Đưa về dạng One-Hot Encoder

dummies = pd.get_dummies(df3['variety'])
```

```python
# Nối các trường dữ liệu

df3 = pd.concat([df3, dummies], axis = 1)
df3.drop('variety', axis = 1, inplace = True)
df3
```

Rời rạc hóa thuộc tính sepal.length

```python
df4 = df3.copy()

df4['sepal.length'] = pd.cut(df4['sepal.length'],
             bins = [-0.1, 0.2, 0.4, 0.6, 0.8, 1],
             labels = ["Very Low", "Low", "Medium", "High", "Very High"])
df4
```

Lấy mẫu ngẫu nhiên có lặp lại 50% dữ liệu

```python
data_sample = df4.sample(frac = 0.5, replace = True)
data_sample
```

Xây dựng mô hình hồi quy tuyến tính để đoán kết quả của thuộc tính variety  dựa trên các thuộc tính khác

```python
df5 = df2.copy()

# Sử dụng 90% dữ liệu để Train và 10% dữ liệu để Test

tp = int(0.9*len(df5))

Train = df5.drop("variety", axis = 1)
Test = df5['variety']

X_train, Y_train = Train[:tp], Test[:tp]
X_test, Y_test = Train[tp:], Test[tp:]
```

```python
from sklearn.linear_model import RidgeClassifier

reg = RidgeClassifier().fit(X_train, Y_train)
pred = reg.predict(X_test)
```

```python
reg.score(X_test, Y_test)
```

```python
target_class = pd.DataFrame({'Test value' : pred, 'True value' : Y_test})
target_class
```

Sử dụng PCA và t-SNE để giảm chiều dữ liệu xuống 2 chiều và hiển thị dữ liệu trong không gian 2 chiều

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
df6 = df2.copy()

finalDf = df6.drop("variety", axis=1)

# Sử dụng PCA để giảm chiều dữ liệu xuống 2 chiều
pca = PCA(n_components=2)
pca_result = pca.fit_transform(finalDf)

# Sử dụng t-SNE để giảm chiều dữ liệu xuống 2 chiều
tsne = TSNE(n_components=2)
tsne_result = tsne.fit_transform(finalDf)

# Hiển thị dữ liệu sau khi giảm chiều
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(pca_result[:, 0], pca_result[:, 1])
plt.title('PCA')

plt.subplot(1, 2, 2)
plt.scatter(tsne_result[:, 0], tsne_result[:, 1])
plt.title('t-SNE')

plt.tight_layout()
plt.show()

```

```python

from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

column = ['Feature1', 'Feature2']
data = pd.DataFrame(pca_result, columns=column)

# Chuẩn hóa dữ liệu
scaler = StandardScaler()
X_scaled = scaler.fit_transform(data)

# Phân cụm dữ liệu sử dụng DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Loại bỏ ngoại lai
mask = labels != -1  # Ngoại lai sẽ được gán nhãn -1
X_clustered = data[mask]

# Hiển thị kết quả phân cụm
plt.scatter(X_clustered['Feature1'], X_clustered['Feature2'], c=labels[mask])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('DBSCAN Clustering - PCA')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

column = ['Feature1', 'Feature2']
data1 = pd.DataFrame(tsne_result, columns=column)

# Chuẩn hóa dữ liệu
scaler = StandardScaler()
X_scaled = scaler.fit_transform(data1)

# Phân cụm dữ liệu sử dụng DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Loại bỏ ngoại lai
mask = labels != -1  # Ngoại lai sẽ được gán nhãn -1
X_clustered = data1[mask]

# Hiển thị kết quả phân cụm
plt.scatter(X_clustered['Feature1'], X_clustered['Feature2'], c=labels[mask])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('DBSCAN Clustering - TSNE')
plt.show()
```
